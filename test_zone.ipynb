```python
# selenium 4
from selenium import webdriver
from selenium.webdriver.firefox.service import Service as FirefoxService
from webdriver_manager.firefox import GeckoDriverManager
from selenium.webdriver.firefox.options import Options



driver = webdriver.Firefox(service=FirefoxService(GeckoDriverManager().install()))
```

```python

print(channel_link)
driver.get(channel_link) # Opens Chrome and go to the given link, Nice :)

# Go To link >>> Inspect >>> copy x path
req_class = 'style-scope ytd-c4-tabbed-header-renderer' # our required info is in this class
xp_name = '//*[@id="text"]'
xp_sub = '//*[@id="subscriber-count"]'
xp_desc = '//*[@id="endpoint"]'
xp_no_vid = '//*[@id="videos-count"]/span[1]'

channel_data = driver.find_element("class name", req_class)
channel_name = channel_data.find_element("xpath", xp_name).text
channel_sub = channel_data.find_element("xpath", xp_sub).text
channel_desc = channel_data.find_element("xpath", xp_desc).text
channel_no_vid = channel_data.find_element("xpath", xp_no_vid).text

# xp_content = '//*[@id="content"]' # giving all of the text data from page - can we use reggex?
# channel_content = channel_data.find_element("xpath", xp_content).text
```
